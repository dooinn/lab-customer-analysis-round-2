```python
display(df['state'].value_counts())
display(df['response'].value_counts())
display(df['vehicle_class'].value_counts())
display(df['vehicle_size'].value_counts())
display(df['vehicle_type'].value_counts())
```

```python
df['state'] = df['state'].fillna('Other')
df['response'] = df['response'].fillna('Other')
df['vehicle_class'] = df['vehicle_class'].fillna('Other')
df['vehicle_size'] = df['vehicle_size'].fillna('Other')
df['vehicle_type'] = df['vehicle_type'].fillna('Other')
```

```python
claim_median = df["months_since_last_claim"].median()
df["months_since_last_claim"] = df["months_since_last_claim"].fillna(claim_median)

complain_median = df["number_of_open_complaints"].median()
df["number_of_open_complaints"] = df["number_of_open_complaints"].fillna(complain_median)
```

## 7. BONUS: Put all the previously mentioned data transformations into a function.

```python
def standard_header(x):
    return x.lower().replace(' ', '_')

df.columns = [standard_header(c) for c in df.columns]
```

```python
def first_quarter(x):
    return (x['month'] == 1) | (x['month'] == 2) | (x['month'] == 3)


df[df.apply(first_quarter, axis=1)]
```
